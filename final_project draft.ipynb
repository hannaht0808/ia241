## 5.2 Background Information on COVID-19
In December of 2019, the first case of COVID-19 was found in Wuhan, China and on January 30, 2020 it was declared a public health emergency of international concern (PHEIC) by the World Health Organization.  COVID-19 is a respiratory disease that is spread person to person mainly through respiratory droplets from an infected person’s sneezes, coughs, and talks. Individuals with underlying conditions or older than 65 years of age are at a higher risk of contracting the virus and for severe illness.  Since COVID-19 was discovered, there have been 138.2 million cases and 2.9 million deaths globally. 

There have been serious implications of the pandemic for millions of people worldwide including unemployment, food shortages, poverty, increased mental health issues, lack of access to proper healthcare, and loss of loved ones.  Nearly half of the 3.3 billion global workforce is at risk of losing their livelihoods and tens of millions of people risk falling into extreme poverty. Border closures, trade restrictions, and lockdown measures implemented since the pandemic have greatly affected the food system and have left low-income countries, marginalized populations, and small-scale farmers at the greatest risk of food scarcity.  

Among the countries affected, Brazil has been hit particularly hard and is having major difficulties managing the pandemic. Brazil has the third most cases following the United States at number one and India at number two.  

The arrival of COVID-19 in Brazil began a vicious cycle of increasing infection and death rates. On February 26, 2020, the first COVID-19 case in Brazil was recorded after a 61-year-old man traveled to the Lombardy region of Italy, an area that had reported a high number of cases and deaths.  Since its arrival in Brazil, there have been over 13.9 million confirmed case and over 371,000 reported deaths.  Brazil’s population consists mainly of young adults and several of its citizens have comorbidities and co-infections that make the population at greater risk for severe health implications caused by the virus.

## Brazil COVID-19 Data

```python
brazil_data = df.loc[ df['countriesAndTerritories']=='Brazil']
brazil_data[:]
```

## Question 1: Which month had the most deaths in 2020 in Brazil? 

```python
brazil_death_sum= brazil_data.groupby('month').sum()['deaths']# this should be brazil_data

brazil_death_sum.plot.bar()
plt.ylabel('deaths')
plt.title('COVID-19 Deaths in Brazil by Month in 2020')
plt.show()
```

## Question 2: What is the trend of confirmed cases in Brazil as of December 2020?

```python
brazil_sum_cases_data= brazil_data.groupby('dateRep').sum()['cases'] # this should be brazil_data
brazil_sum_cases_data.plot()
plt.ylabel('# of cases')
plt.title('Trend of Confirmed COVID-19 Cases in Brazil 2020')
plt.show()
```

## Question 3: How the # deaths is related to the # cases in Brazil?

```python
brazil_data = df.loc[ df['countriesAndTerritories']=='Brazil']
brazil_data.plot.scatter(x='cases',y='deaths',c='month')
plt.title('How the # deaths is related to the # cases in Brazil in 2020')
plt.xlabel('cases')
plt.show()
```
